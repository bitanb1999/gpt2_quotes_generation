```
import gdown
# https://drive.google.com/file/d/1lyo7edhxQfimeXq45kjsp8y9XuXs9gej/view?usp=sharing
url = "https://drive.google.com/uc?id=1lyo7edhxQfimeXq45kjsp8y9XuXs9gej"
output = "quotes.csv"
gdown.download(url, output, quiet=False)
```

```
import pandas as pd
```

```
quotes_df = pd.read_csv(output)
quotes_df.head()
```

```
quotes_df.columns
```

```
remove_columns = [x for x in quotes_df.columns if 'Unnamed: ' in x]
remove_columns
```

```
quotes_df.drop(columns=remove_columns, inplace=True)
quotes_df.columns
```

```
quotes_df.tail(1)
```

```
add_row = quotes_df.columns.tolist()
```

```
quotes_df.loc[quotes_df.shape[0], :] = add_row
```

```
quotes_df.tail(2)
```

```
quotes_df.columns = ['quote', 'author', 'genre']
```

```
quotes_df.head()
```

```
quotes_df.to_csv('processed_quotes.csv', index=False)
```

```
from sklearn.model_selection import train_test_split
import re
```

```
train_test_ratio = 0.9
train_valid_ratio = 7/9
df_full_train, df_test = train_test_split(quotes_df, train_size = train_test_ratio, random_state = 1)
df_train, df_valid = train_test_split(df_full_train, train_size = train_valid_ratio, random_state = 1)
```

```
print(df_train.shape)
print(df_test.shape)
print(df_valid.shape)
```

```
def process_data(df):
  data = []
  for quote in df['quote'].tolist():
    quote = str(quote).strip()
    quote = re.sub(r"\s", " ", quote)
    tmp = ['<BOS>', quote, '<EOS>']
    data.append(' '.join(tmp))
  return data
```

```
processed_val = process_data(df_valid)
processed_val[-1]
```

```
processed_train = process_data(df_train)
processed_train[0]
```

```
processed_test = process_data(df_test)
processed_test[0]
```

```
def generate_text_file(contents, filename):
  with open(filename, 'w', encoding='utf-8') as fhead:
    for content in contents:
      fhead.write(content+'\n')
```

```
generate_text_file(processed_train, 'train.txt')
```

```
generate_text_file(processed_val, 'val.txt')
```

```
generate_text_file(processed_test, 'test.txt')
```
